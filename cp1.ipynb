```python

survey_new = pd.read_csv('./CP1 Stack Overflow Database.csv', usecols=['RemoteWork'])

hybrid_count = (survey_new['RemoteWork'] == 'Hybrid').sum()
remote_count = (survey_new['RemoteWork'] == 'Remote').sum()
office_count = (survey_new['RemoteWork'] == 'Office').sum()

total_respondents = len(survey_new)
percentage_hybrid = (hybrid_count / total_respondents) * 100
percentage_remote = (remote_count / total_respondents) * 100
percentage_office = (office_count / total_respondents) * 100

print(f"Porcentagem de respondentes que trabalham no formato híbrido: {percentage_hybrid:.2f}%")
print(f"Porcentagem de respondentes que trabalham no formato remoto: {percentage_remote:.2f}%")
print(f"Porcentagem de respondentes que trabalham no formato presencial: {percentage_office:.2f}%")
#Porcentagem de respondentes que trabalham no formato híbrido: 0.00%
#Porcentagem de respondentes que trabalham no formato remoto: 35.83%
#Porcentagem de respondentes que trabalham no formato presencial: 0.00%
```

```python

survey_new = pd.read_csv('./CP1 Stack Overflow Database.csv', usecols=['Country', 'ConvertedCompYearly'])

countries_of_interest = ['United States of America', 'United Kingdom of Great Britain and Northern Ireland', 'Canada']
survey_country = survey_new[survey_new['Country'].isin(countries_of_interest)].copy()

country_stats = survey_country.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'std'])

num_observations = survey_country.shape[0]

country_max_mean = country_stats['mean'].idxmax()
max_mean_salary = country_stats.loc[country_max_mean, 'mean']

country_max_std = country_stats['std'].idxmax()
max_std_salary = country_stats.loc[country_max_std, 'std']

print(f"Número de observações no DataFrame 'survey_country': {num_observations}")
print(f"País com a maior média de salários: {country_max_mean} (Média: {max_mean_salary:.2f})")
print(f"País com o maior desvio-padrão entre os salários: {country_max_std} (Desvio-padrão: {max_std_salary:.2f})")

#Número de observações no DataFrame 'survey_country': 18889
#País com a maior média de salários: United States of America (Média: 174645.47)
#País com o maior desvio-padrão entre os salários: Canada (Desvio-padrão: 1922107.55)
```

```python

survey_new = pd.read_csv('./CP1 Stack Overflow Database.csv', usecols=['RemoteWork'])
survey_country = pd.read_csv('./CP1 Stack Overflow Database.csv', usecols=['Country', 'RemoteWork'])
countries_of_interest = ['United States of America', 'United Kingdom of Great Britain and Northern Ireland', 'Canada']
survey_country = survey_country[survey_country['Country'].isin(countries_of_interest)].copy()
percentage_hybrid_country = (survey_country['RemoteWork'] == 'Hybrid').mean() * 100
percentage_remote_country = (survey_country['RemoteWork'] == 'Remote').mean() * 100
percentage_office_country = (survey_country['RemoteWork'] == 'Office').mean() * 100
print("Porcentagem de respondentes que trabalham no formato híbrido, remoto e presencial no DataFrame 'survey_country':")
print(f"Formato Híbrido: {percentage_hybrid_country:.2f}%")
print(f"Formato Remoto: {percentage_remote_country:.2f}%")
print(f"Formato Presencial: {percentage_office_country:.2f}%")
percentage_remote_new = (survey_new['RemoteWork'] == 'Remote').mean() * 100
print(f"\nPorcentagem de respondentes que trabalham no formato remoto no DataFrame 'survey_new': {percentage_remote_new:.2f}%")
if percentage_remote_new > percentage_remote_country:
    print("\nHá mais trabalhadores remotos no DataFrame principal 'survey_new'.")
elif percentage_remote_new < percentage_remote_country:
    print("\nHá mais trabalhadores remotos nos países selecionados no novo DataFrame 'survey_country'.")
else:
    print("\nA quantidade de trabalhadores remotos é a mesma nos dois DataFrames.")
#Porcentagem de respondentes que trabalham no formato híbrido, remoto e presencial no DataFrame 'survey_country':
#Formato Híbrido: 0.00%
#Formato Remoto: 47.72%
#Formato Presencial: 0.00%

#Porcentagem de respondentes que trabalham no formato remoto no DataFrame 'survey_new': 35.83%

#Há mais trabalhadores remotos nos países selecionados no novo DataFrame 'survey_country'.
```

```python
survey_country = pd.read_csv('./CP1 Stack Overflow Database.csv', usecols=['Country', 'Industry', 'ConvertedCompYearly'])
industries_of_interest = ['Information Services, IT, Software Development, or other Technology', 'Financial Services', 'Insurance']
survey_country_ind = survey_country[survey_country['Industry'].isin(industries_of_interest)].copy()
country_ind_mean_salary = survey_country_ind.groupby(['Country', 'Industry'])['ConvertedCompYearly'].mean()
max_mean_salary_industry = country_ind_mean_salary.idxmax()
min_mean_salary_industry = country_ind_mean_salary.idxmin()
print("Indústria de cada país com a maior média salarial:")
print(max_mean_salary_industry)
print("\nIndústria de cada país com a menor média salarial:")
print(min_mean_salary_industry)
#Indústria de cada país com a maior média salarial:
#('Thailand', 'Financial Services')
#Indústria de cada país com a menor média salarial:
#('Rwanda', 'Information Services, IT, Software Development, or other Technology')

```

```python
survey_country_ind = pd.read_csv('./CP1 Stack Overflow Database.csv', usecols=['Industry', 'WorkExp'])
industries_of_interest = ['Information Services, IT, Software Development, or other Technology', 'Financial Services', 'Insurance']
survey_country_ind = survey_country_ind[survey_country_ind['Industry'].isin(industries_of_interest)].copy()
industry_mean_workexp = survey_country_ind.groupby('Industry')['WorkExp'].mean()
max_mean_workexp_industry = industry_mean_workexp.idxmax()
max_mean_workexp = industry_mean_workexp.max()
min_mean_workexp_industry = industry_mean_workexp.idxmin()
min_mean_workexp = industry_mean_workexp.min()

print("Média de anos de experiência ('WorkExp') para cada indústria:")
print(industry_mean_workexp)
print("\nIndústria com os desenvolvedores com mais experiência:")
print(f"{max_mean_workexp_industry} (Média de anos de experiência: {max_mean_workexp:.2f})")
print("\nIndústria com os desenvolvedores com menos experiência:")
print(f"{min_mean_workexp_industry} (Média de anos de experiência: {min_mean_workexp:.2f})")

#Média de anos de experiência ('WorkExp') para cada indústria:
#Industry
#Financial Services                                                     11.615722
#Information Services, IT, Software Development, or other Technology    11.420499
#Insurance                                                              12.416996
#Name: WorkExp, dtype: float64
#Indústria com os desenvolvedores com mais experiência:
#Insurance (Média de anos de experiência: 12.42)
#Indústria com os desenvolvedores com menos experiência:
#Information Services, IT, Software Development, or other Technology (Média de anos de experiência: 11.42)

```
